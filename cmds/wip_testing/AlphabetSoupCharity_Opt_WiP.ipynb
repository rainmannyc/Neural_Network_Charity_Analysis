### Deliverable 1: Preprocessing the Data for a Neural Network

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("../Resources/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns=['NAME', 'EIN'])
```

```python
# Determine the number of unique values in each column
application_df.nunique()
```

```python
application_df["APPLICATION_TYPE"]
```

```python
application_df.APPLICATION_TYPE.value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_types = application_df.APPLICATION_TYPE.value_counts()
```

```python
# Visualize the value counts of APPLICATION_TYPE
app_types.plot.density()
```

```python
# Determine which values to replace if counts are less than ...?
replace_application = list(app_types[app_types < 500 ].index)

# Replace in dataframe
for app in replace_application:
    application_df.APPLICATION_TYPE = application_df.APPLICATION_TYPE.replace(app,"Other")
    
# Check to make sure binning was successful
application_df.APPLICATION_TYPE.value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
classtypes = application_df.CLASSIFICATION.value_counts()
```

```python
# Visualize the value counts of CLASSIFICATION
classtypes.plot.density()
```

```python
# Determine which values to replace if counts are less than ..?
replace_class = list(classtypes[classtypes < 1000].index)

# Replace in dataframe
for cls in replace_class:
    application_df.CLASSIFICATION = application_df.CLASSIFICATION.replace(cls,"Other")
    
# Check to make sure binning was successful
application_df.CLASSIFICATION.value_counts()
```

```python
# Generate our categorical variable lists
application_cat = application_df.dtypes[application_df.dtypes == "object"].index.tolist()
```

```python
application_cat
```

```python
application_df[application_cat].nunique()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# # Fit and transform the OneHotEncoder using the categorical variable list

# success = pd.DataFrame(enc.fit_transform(application_df.IS_SUCCESSFUL.values.reshape(-1,1)))
encode_df = pd.DataFrame(enc.fit_transform(application_df[application_cat]))


# encode_df = pd.DataFrame(enc.fit_transform(ramen_df.Country.values.reshape(-1,1)))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(application_cat)
encode_df.columns
```

```python
# Merge one-hot encoded features and drop the originals
# application_df = application_df.append(success, ignore_index=True)
application_df = application_df.merge(encode_df,left_index=True, right_index=True)

application_df = application_df.drop(application_cat,1)
application_df.head()
```

```python
# application_df.columns
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df["IS_SUCCESSFUL"].values
X = application_df.drop(["IS_SUCCESSFUL","SPECIAL_CONSIDERATIONS_Y", "SPECIAL_CONSIDERATIONS_N", "CLASSIFICATION_Other", 'CLASSIFICATION_C1000', 'CLASSIFICATION_C1200',
       'CLASSIFICATION_C2000','CLASSIFICATION_C2100', 'CLASSIFICATION_C3000'],1).values

X
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=24)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

### Deliverable 2: Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 = 600
hidden_nodes_layer2 = 400
hidden_nodes_layer3 = 300
hidden_nodes_layer4 = 250
hidden_nodes_layer5 = 100
hidden_nodes_layer6 = 60




nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="tanh")
)

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="LeakyReLU"))

# Third hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer3, activation="relu"))

# Fourth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer4, activation="relu"))

# Fifth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer5, activation="relu"))

# Sixth hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer6, activation="linear"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Import checkpoint dependencies
import os
from tensorflow.keras.callbacks import ModelCheckpoint

# Define the checkpoint path and filenames
os.makedirs("checkpoints/",exist_ok=True)
checkpoint_path = "checkpoints/weights.{epoch:02d}.hdf5"
```

```python
# opt = tf.keras.optimizers.Adagrad(learning_rate=0.03)
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="sgd", metrics=["accuracy"])
```

```python
# Create a callback that saves the model's weights every epoch
steps_per_execution=4000

cp_callback = ModelCheckpoint(
    filepath=checkpoint_path,
    verbose=1,
    save_weights_only=True,
    save_freq= steps_per_execution,
    )
```

```python
# Train the model

fit_model = nn.fit(X_train_scaled,y_train,epochs=100,callbacks=[cp_callback])
```

```python

plt.plot(fit_model.history["loss"])
plt.title("Loss - 6 Hidden Layers")
plt.legend(["scaled"])
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
